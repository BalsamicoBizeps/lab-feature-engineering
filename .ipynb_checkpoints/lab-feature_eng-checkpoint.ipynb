```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

```python
spaceship.dtypes
```

**Check for missing values**

```python
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship = spaceship.dropna(how="any")
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship["Cabin"] = spaceship["Cabin"].apply(lambda x: x.split("/")).apply(lambda x: x[0])
spaceship["Cabin"].value_counts()
```

- Drop PassengerId and Name

```python
spaceship.drop(columns = ["PassengerId", "Name"], inplace = True)
```

- For non-numerical columns, do dummies.

```python
spaceship[["CryoSleep", "VIP"]] = spaceship[["CryoSleep", "VIP"]].applymap(lambda x: True if x == "True" else False)
```

```python
cat_df = spaceship.select_dtypes("object")
num_df = spaceship.select_dtypes("number")
bool_df = spaceship.select_dtypes("bool")
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
df_cat_num = pd.get_dummies(cat_df, dtype = int)
```

```python
enc = OneHotEncoder()
```

```python
enc.fit(cat_df)
```

```python
enc.transform(cat_df).toarray()
```

```python
spaceship_num = pd.concat([df_cat_num, num_df, bool_df.drop(columns = "Transported")], axis=1)
spaceship_num.columns
```

```python
spaceship_corr = spaceship_num.corr()
```

```python
plt.figure(figsize=(10, 8))  # Set the size of the figure
sns.heatmap(spaceship_corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

**Perform Train Test Split**

```python
features = spaceship_num
target = spaceship["Transported"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 0)
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(X_train_scaled, y_train)
```

- Evaluate your model's performance. Comment it

```python
knn.score(X_test_scaled, y_test)
```
